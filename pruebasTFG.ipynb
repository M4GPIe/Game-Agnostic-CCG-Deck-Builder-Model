```python
from featureExtractor.FeatureExtractor import FeatureExtractor
from inputFiles.effects import effect_references
from inputFiles.keywords import keywords_data
from featureExtractor.utils.supportedModels import SupportedModels

raw_cards = 'inputFiles/RawCards.json.zip'
supported_cards = 'inputFiles/supportedCards.txt'
output_file = 'outputFiles/newParsedCards.json'

feature_extractor = FeatureExtractor(
    nlp_model = SupportedModels.DISTILBERT_BASE_NLI_STSB_MEAN_TOKENS.value, 
    nlp_threshold = 0.6, 
    verbose = False, 
    chunkSize = 500, 
    keywords_list = keywords_data, 
    effect_references= effect_references, 
    excluded_card_types = ["Land"], 
    supported_cards_file= supported_cards
)

feature_extractor.parse_cards(input_zip_file = raw_cards, output_file = output_file)

```
