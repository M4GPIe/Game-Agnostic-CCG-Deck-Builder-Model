```python
import zipfile
import json
from utils.helper_functions import parse_card_data

# file containing all cards
zip_path = 'content\AtomicCards.json.zip'
 
# output file containing all parsed cards
output_file = 'content\parsedCards.json'

with zipfile.ZipFile(zip_path, 'r') as zip_ref:
    json_files = [f for f in zip_ref.namelist() if f.endswith('.json')]
    if not json_files:
        print("No se encontró ningún archivo JSON en el zip.")
    else:
        json_file = json_files[0]
        print("Procesando el archivo:", json_file)
        with zip_ref.open(json_file) as f:
            data = json.load(f)

            cards = list(data['data'].values())
            parsed_cards = []

            for card_list in cards:
                print(f"Processing {len(parsed_cards)} out of {len(cards)}")
                card = card_list[0]
                if card.get("type") == "Land":
                    continue
                parsed_card = parse_card_data(card)  
                if parsed_card:  
                    parsed_cards.append(parsed_card)

            #display_cards_as_table(parsed_cards)

        with open(output_file, 'w') as json_out:
            json.dump([card.__dict__ for card in parsed_cards], json_out, indent=4)

        print(f"Cartas guardadas en {output_file}")
```
